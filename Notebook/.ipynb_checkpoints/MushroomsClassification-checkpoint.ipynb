# Mushrooms Classification
Mushrooms classification as edible or poisonouns is an important problem for many poeple. The classification process can be done using different mushrooms features. This notebook is intended to solve this problem using machine learning algorithms that can predict the class of each mushroom (edible or poisonous) from its features. The dataset used for this project is: [UCI Machine Learning Repository: Mushroom Data Set](https://archive.ics.uci.edu/ml/datasets/mushroom).

6. Evaluation.
7. Validation.

## 4. Building The Benchmark Model

```python
def train(model, X_train, y_train):
    '''
    This functions trains the given model
    on the training varaiables.
    '''
    results = {}
    start = time() 
    model = model.fit(X_train, y_train)
    end = time() 
    
    results['train_time'] = end - start

    print("{} trained succesfully.".format(model.__class__.__name__))
    return results
```

## 6. Evaluation

```python
def predict(trained_model, X_train, X_test, y_train, y_test):
    '''
    This functions outputs the
    accuracy score for training and testing.
    '''
    results = {}

    start = time() 
    y_pred_test = trained_model.predict(X_test)
    y_pred_train = trained_model.predict(X_train)
    end = time() 
    
    # The total prediction time
    results['pred_time'] = end - start
            
    # Training accuracy
    results['acc_train'] = accuracy_score(y_train, y_pred_train)
        
    # Testing accuracy
    results['acc_test'] = accuracy_score(y_test, y_pred_test)
    
    # F-score 0.5 on training data
    results['f_train'] = fbeta_score(y_pred_train, y_train, beta=0.5)
        
    # F-score 0.5 on training data
    results['f_test'] = fbeta_score(y_pred_test, y_test, beta=0.5)
       
    # F-score 0.5 on training data
    results['recall_train'] = recall_score(y_pred_train, y_train)
        
    # F-score 0.5 on training data
    results['recall_test'] = recall_score(y_pred_test, y_test)
    
    # F-score 0.5 on training data
    results['precision_train'] = precision_score(y_pred_train, y_train)
        
    # F-score 0.5 on training data
    results['precision_test'] = precision_score(y_pred_test, y_test)
    
        
    # Success
    print("{} evaluation metrics has been recorded succesfully.".format(model.__class__.__name__))
        
#     print("Testing on training data:")
#     print(classification_report(y_train, y_pred_train))
#     print("Testing on testing data:")
#     print(classification_report(y_test, y_pred_test))
    
    # Return the results
    return results
    
```

```python
for model in [NB_model, RF_model, SVM_model, LR_model]:
    model_name = model.__class__.__name__
    results[model_name].update(predict(model, X_train, X_test, y_train, y_test))
    
```

```python
# display(results)
results_df = pd.DataFrame(results)
results_df = results_df.T
display(results_df)
```

## 7. Validation

```python
from sklearn.grid_search import GridSearchCV

parameters = {
    'n_estimators': [5, 10, 15, 25, 50, 100],
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 50, 100],
    'min_samples_split': [2,5,10],
    'min_samples_leaf': [1, 5, 10, 50],
    'max_features': ['auto', 'sqrt', 'log2', None]
}

# scorer = make_scorer(fbeta_score, beta=0.5)
grid_obj = GridSearchCV(RF_model, param_grid=parameters, scoring='accuracy')
grid_fit = grid_obj.fit(X_train, y_train)
best_clf = grid_fit.best_estimator_
```

```python
predictions = (RF_model.fit(X_train, y_train)).predict(X_test)
best_predictions = best_clf.predict(X_test)

# Report the before-and-afterscores
print("Unoptimized model\n------")
print("Accuracy score on testing data: {:.4f}".format(accuracy_score(y_test, predictions)))
print("F-score on testing data: {:.4f}".format(fbeta_score(y_test, predictions, beta = 0.5)))
print("\nOptimized Model\n------")
print("Final accuracy score on the testing data: {:.4f}".format(accuracy_score(y_test, best_predictions)))
print("Final F-score on the testing data: {:.4f}".format(fbeta_score(y_test, best_predictions, beta = 0.5)))

```
